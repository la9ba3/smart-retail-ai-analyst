```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
monthly_sales = pd.read_csv("../data/processed/monthly_sales.csv")
```

```python
monthly_sales
```

```python
plt.figure(figsize=(12,5))
plt.plot(monthly_sales["YearMonth"],monthly_sales["total_revenue"],marker="o")
plt.title("Monthly revenue - Online Retail")
plt.xlabel("Month")
plt.xlabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Conclusion

L'analyse temporelle permet d'observer l'évolution du chiffre d'affaires par mois.

Le graphique aide à repérer :
- les mois les plus performants ;
- les périodes plus faibles ;
- une possible saisonnalité ;
- les tendances globales des ventes.

Ces résultats pourront être approfondis avec une analyse par produit, pays ou segment client.
