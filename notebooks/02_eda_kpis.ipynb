```python
import pandas as pd 
```

```python
df = pd.read_csv("../data/processed/online_retail_clean.csv")
```

```python
total_revenue = df["TotalPrice"].sum()
total_invoices = df["InvoiceNo"].nunique()
total_customers = df["CustomerID"].nunique()
total_products = df["StockCode"].nunique()
total_countries = df["Country"].nunique()
average_basket = total_revenue / total_invoices

kpis = {
    "total_revenue": total_revenue,
    "total_invoices": total_invoices,
    "total_customers": total_customers,
    "total_products": total_products,
    "total_countries": total_countries,
    "average_basket": average_basket,
}

kpis
```

## Conclusion

Les KPIs globaux donnent une première vue d'ensemble de l'activité retail.

Ils permettent de comprendre rapidement :
- le chiffre d'affaires total généré ;
- le volume de commandes ;
- la taille de la base client ;
- la diversité des produits vendus ;
- la couverture géographique ;
- le panier moyen par facture.

Ces indicateurs serviront de base pour les prochaines analyses business.
